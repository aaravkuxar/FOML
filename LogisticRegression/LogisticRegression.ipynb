```python
import pandas as pd
social_media_analytics = pd.read_csv('/home/aaravk/Documents/FOML/Datasets/social_media_analytics.csv')
social_media_analytics
```

```python
social_media_analytics = social_media_analytics[['income','isHomeOwner']]
social_media_analytics
```

```python
social_media_analytics.isHomeOwner = social_media_analytics.isHomeOwner.replace({True: 1, False: 0})
social_media_analytics
```

```python
import numpy as np
x = np.array(social_media_analytics['income']).reshape(-1,1)
y = np.array(social_media_analytics['isHomeOwner']).reshape(-1,1)
```

```python
from matplotlib import pyplot as plt
import seaborn as sns
sns.regplot(x = social_media_analytics['income'], y = social_media_analytics['isHomeOwner'], data = social_media_analytics, logistic = True, color = 'blue')
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test = train_test_split(x, y, test_size = 0.1)
```

```python
from sklearn.linear_model import LogisticRegression
model = LogisticRegression()
model.fit(x_train,y_train)
```

```python
predicted_values = model.predict(x_test)
predicted_values = np.array(predicted_values).reshape(-1,1)
print(predicted_values)
modelScore = model.score(x_test,y_test)
print("Your Model Score: ")
modelScore
```
