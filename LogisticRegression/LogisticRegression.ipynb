```python
import pandas as pd
heart_disease = pd.read_csv('/home/aaravk/Documents/FOML/Datasets/heart_disease.csv')
heart_disease
```

```python
heart_disease = heart_disease[['age','prevalentHyp']]
heart_disease
```

```python
import numpy as np
x = np.array(heart_disease['age']).reshape(-1,1)
y = np.array(heart_disease['prevalentHyp']).reshape(-1,1)
```

```python
from matplotlib import pyplot as plt
import seaborn as sns
sns.regplot(x = heart_disease['age'], y = heart_disease['prevalentHyp'], data = heart_disease, logistic = True, color = 'blue')
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test = train_test_split(x, y, test_size = 0.1)
```

```python
from sklearn.linear_model import LogisticRegression
model = LogisticRegression()
model.fit(x_train,y_train)
```

```python
predicted_values = model.predict(x_test)
predicted_values = np.array(predicted_values).reshape(-1,1)
print(predicted_values)
modelScore = model.score(x_test,y_test)
print("Your Model Score: ")
modelScore
```
